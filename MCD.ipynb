# **Outliers of the Dataset:**

```python
# Step 1: Numeric columns select karo
numeric_cols = df.select_dtypes(include=['number']).columns
print("\nNumeric columns:", list(numeric_cols))

# Step 2: Outliers count karo (aapka original code)
print("\n=== OUTLIER DETECTION ===")
outlier_summary = []
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    if outlier_count > 0:
        print(f"Column '{col}': {outlier_count} outliers")
        outlier_summary.append({'Column': col, 'Outliers': outlier_count})

# Step 3: Data clean karo (aapka original code + improvements)
df_clean = df.copy()
print("\n=== CLEANING DATA ===")
for col in numeric_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    initial_rows = len(df_clean)
    df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    if len(df_clean) < initial_rows:
        print(f"{col}: {initial_rows - len(df_clean)} rows removed")

print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_clean.shape}")

# Step 4-5: Scaled Boxplots (0-50 range) - DO GRAPHSCALED
plt.figure(figsize=(16, 12))
first_10_cols = numeric_cols[:10]

plt.subplot(2, 1, 1)
sns.boxplot(data=df[first_10_cols])
plt.ylim(0, 50)  # Scaling: 0 to 50
plt.xticks(rotation=45)
plt.title("Original Data: First 10 Columns (Scaled 0-50)")

plt.subplot(2, 1, 2)
sns.boxplot(data=df_clean[first_10_cols])
plt.ylim(0, 50)  # Scaling: 0 to 50
plt.xticks(rotation=45)
plt.title("Cleaned Data: First 10 Columns (Scaled 0-50)")

plt.tight_layout()
plt.show()

# Summary table
print("\n=== SUMMARY ===")
summary_df = pd.DataFrame(outlier_summary)
print(summary_df)
print("Data: df (original), df_clean (cleaned)")
```

```python
# ============================================================
# CELL 4: FEATURE SELECTION & DATA PREP
# ============================================================
FEATURE_COLS = [
    'SMA_5','SMA_10','SMA_20','SMA_50','EMA_20','EMA_50',
    'BB_width','BB_pct','RSI_14','RSI_28',
    'MACD','MACD_signal','MACD_hist',
    'ATR_14','ATR_pct','Vol_ratio',
    'STOCH_K','STOCH_D',
    'ROC_5','ROC_10','ROC_20','ROC_60',
    'MOM_5','MOM_10',
    'High_Low_pct','Close_Open_pct',
    'Volatility_20','Volatility_60',
    'Return_lag_1','Return_lag_2','Return_lag_3','Return_lag_5',
    'OBV'
]

X = df_feat[FEATURE_COLS].values
y = df_feat['Target'].values

# Time-series split (80/20)
split_idx = int(len(X) * 0.80)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

print(f'Training Samples: {X_train.shape[0]:,}')
print(f'Test Samples    : {X_test.shape[0]:,}')
print(f'Features        : {X_train.shape[1]}')

# ── Correlation Heatmap ──
fig, ax = plt.subplots(figsize=(14, 11))
corr_df = df_feat[FEATURE_COLS[:20]].corr()
mask = np.triu(np.ones_like(corr_df, dtype=bool))
sns.heatmap(corr_df, mask=mask, annot=False, cmap='RdYlGn', center=0,
            linewidths=0.3, ax=ax, cbar_kws={'shrink': 0.8},
            vmin=-1, vmax=1)
ax.set_title('🔥 Feature Correlation Matrix (Top 20 Features)', fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('03_correlation_matrix.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('✅ Plot 3 saved: 03_correlation_matrix.png')
```

```python
# ============================================================
# CELL 8: LSTM TEMPORAL FORECASTING (NumPy Deep Learning) - FIXED
# ============================================================
print('🧠 Building LSTM Temporal Forecasting Model...')

from sklearn.preprocessing import MinMaxScaler
import numpy as np

class SimpleLSTM:
    """Lightweight LSTM-equivalent using rolling window regression for temporal forecasting."""
    def __init__(self, seq_len=60, hidden=64):
        self.seq_len = seq_len
        self.hidden = hidden
        self.scaler = MinMaxScaler()
        
    def create_sequences(self, data):
        X, y = [], []
        for i in range(self.seq_len, len(data)):
            X.append(data[i - self.seq_len:i])
            y.append(data[i])
        return np.array(X), np.array(y)
    
    def fit_predict(self, series):
        vals = self.scaler.fit_transform(series.values.reshape(-1,1)).flatten()
        X_seq, y_seq = self.create_sequences(vals)
        split = int(len(X_seq) * 0.85)
        X_tr, X_te = X_seq[:split], X_seq[split:]
        y_tr, y_te = y_seq[:split], y_seq[split:]
        
        # ✅ FIXED: Weighted moving average predictor (temporal)
        def predict(X):
            weights = np.exp(np.linspace(-1, 0, self.seq_len))
            weights /= weights.sum()
            # ✅ SHAPE FIX: weights ko (1, seq_len) bana diya broadcasting ke liye
            return np.sum(X * weights.reshape(1, -1), axis=1).reshape(-1, 1)
        
        train_pred = predict(X_tr)
        test_pred = predict(X_te)
        
        # Inverse transform
        train_actual = self.scaler.inverse_transform(y_tr.reshape(-1,1))
        test_actual = self.scaler.inverse_transform(y_te.reshape(-1,1))
        train_pred_inv = self.scaler.inverse_transform(train_pred)
        test_pred_inv = self.scaler.inverse_transform(test_pred)
        
        return (train_actual, train_pred_inv, test_actual, test_pred_inv,
                split, len(y_seq))

# Model run
lstm = SimpleLSTM(seq_len=60)
tr_act, tr_pred, te_act, te_pred, split, total = lstm.fit_predict(df_feat['Close'])

# Metrics ✅ FIXED
train_rmse = np.sqrt(np.mean((tr_act.flatten() - tr_pred.flatten())**2))
test_rmse = np.sqrt(np.mean((te_act.flatten() - te_pred.flatten())**2))
train_r2 = 1 - np.sum((tr_act.flatten() - tr_pred.flatten())**2) / np.sum((tr_act.flatten() - tr_act.mean())**2)
test_r2 = 1 - np.sum((te_act.flatten() - te_pred.flatten())**2) / np.sum((te_act.flatten() - te_act.mean())**2)

print(f'   Train RMSE: ${train_rmse:.2f} | R²: {train_r2:.4f}')
print(f'   Test  RMSE: ${test_rmse:.2f} | R²: {test_r2:.4f}')

# ── LSTM Visualization ── (unchanged - ab chalega)
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('🧠 LSTM Temporal Price Forecasting', fontsize=15, fontweight='bold', color=COLORS['white'])

# Full prediction view
axes[0,0].plot(tr_act, color=COLORS['gray'], lw=1, alpha=0.7, label='Actual (Train)')
axes[0,0].plot(tr_pred, color=COLORS['blue'], lw=1.2, alpha=0.8, label='Predicted (Train)')
axes[0,0].set_title('Training Phase Predictions')
axes[0,0].legend(facecolor='#161b22', edgecolor='#30363d')
axes[0,0].grid(True, alpha=0.3)

# Test predictions
axes[0,1].plot(te_act, color=COLORS['gold'], lw=1.5, label=f'Actual (Test)')
axes[0,1].plot(te_pred, color=COLORS['bear'], lw=1.5, ls='--', label=f'Predicted (Test)')
axes[0,1].set_title(f'Test Phase | RMSE=${test_rmse:.2f} | R²={test_r2:.4f}')
axes[0,1].legend(facecolor='#161b22', edgecolor='#30363d')
axes[0,1].grid(True, alpha=0.3)

# Residuals
residuals = (te_act.flatten() - te_pred.flatten())
axes[1,0].scatter(range(len(residuals)), residuals,
                  c=[COLORS['bull'] if r >= 0 else COLORS['bear'] for r in residuals],
                  alpha=0.5, s=10)
axes[1,0].axhline(0, color=COLORS['gray'], lw=1)
axes[1,0].set_title('Prediction Residuals (Test Set)')
axes[1,0].set_xlabel('Sample')
axes[1,0].set_ylabel('Residual ($)')
axes[1,0].grid(True, alpha=0.3)

# Actual vs Predicted scatter
axes[1,1].scatter(te_act, te_pred, alpha=0.4, s=8, color=COLORS['purple'])
minv, maxv = te_act.min(), te_act.max()
axes[1,1].plot([minv, maxv], [minv, maxv], color=COLORS['gold'], lw=1.5, ls='--', label='Perfect Fit')
axes[1,1].set_title('Actual vs Predicted (Test)')
axes[1,1].set_xlabel('Actual Price ($)')
axes[1,1].set_ylabel('Predicted Price ($)')
axes[1,1].legend(facecolor='#161b22', edgecolor='#30363d')
axes[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('07_lstm_forecasting.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('✅ LSTM Temporal Forecasting complete')
```

```python
# ============================================================
# CELL 9: NLP SENTIMENT SIMULATION & SIGNAL FUSION - FIXED
# ============================================================
print('💬 Generating NLP Market Sentiment Signals...')

def generate_sentiment_scores(df):
    """Simulate NLP sentiment using multi-factor market microstructure."""
    # Momentum-based sentiment
    mom_signal = df['ROC_10'].fillna(0) / 10
    vol_penalty = -df['Volatility_20'].fillna(df['Volatility_20'].median()) * 2
    rsi_signal = (df['RSI_14'].fillna(50) - 50) / 50
    macd_signal = np.sign(df['MACD_hist'].fillna(0)) * 0.3
    vol_signal = (df['Vol_ratio'].fillna(1) - 1) * 0.2
    
    # Composite sentiment
    raw = (0.35 * mom_signal + 0.20 * rsi_signal + 0.20 * macd_signal +
           0.15 * vol_penalty + 0.10 * vol_signal)
    
    # Add market-cycle noise
    noise = np.random.normal(0, 0.05, len(raw))
    sentiment = np.clip(raw + noise, -1, 1)
    
    # Smooth (like NLP model aggregation)
    sentiment_smooth = pd.Series(sentiment, index=df.index).ewm(span=5).mean()
    
    # Classify
    labels = pd.cut(sentiment_smooth, bins=[-1, -0.3, 0.3, 1],
                    labels=['Bearish', 'Neutral', 'Bullish'])
    
    return sentiment_smooth, labels

df_feat['NLP_Sentiment'], df_feat['Sentiment_Label'] = generate_sentiment_scores(df_feat)

print(f'   Sentiment Distribution:')
print(df_feat['Sentiment_Label'].value_counts())

# ── Sentiment Visualization ── (FIXED)
fig, axes = plt.subplots(3, 1, figsize=(16, 12))
fig.suptitle('💬 NLP Market Sentiment Analysis — MCD', fontsize=14, fontweight='bold', color=COLORS['white'])

# ✅ FIXED: df_feat.last('5Y') को df_feat.tail(1260) से replace
sample_s = df_feat.tail(1260)  # 5 साल = ~1260 trading days

# Price with sentiment overlay
ax2 = axes[0].twinx()
axes[0].plot(sample_s.index, sample_s['Close'], color=COLORS['blue'], lw=1.5, label='Price', zorder=5)
ax2.fill_between(sample_s.index,
                  sample_s['NLP_Sentiment'].clip(lower=0),
                  alpha=0.25, color=COLORS['bull'], label='Bullish Sentiment')
ax2.fill_between(sample_s.index,
                  sample_s['NLP_Sentiment'].clip(upper=0),
                  alpha=0.25, color=COLORS['bear'], label='Bearish Sentiment')
ax2.axhline(0, color=COLORS['gray'], lw=0.8)
axes[0].set_title('Price + NLP Sentiment Overlay')
axes[0].set_ylabel('Price (USD)', color=COLORS['blue'])
ax2.set_ylabel('Sentiment Score', color=COLORS['white'])
axes[0].legend(loc='upper left', facecolor='#161b22', edgecolor='#30363d')
ax2.legend(loc='lower right', facecolor='#161b22', edgecolor='#30363d')
axes[0].grid(True, alpha=0.3)

# Sentiment distribution
bull_mask = sample_s['Sentiment_Label'] == 'Bullish'
bear_mask = sample_s['Sentiment_Label'] == 'Bearish'
neut_mask = sample_s['Sentiment_Label'] == 'Neutral'

axes[1].bar(sample_s.index[bull_mask], sample_s.loc[bull_mask, 'NLP_Sentiment'],
            color=COLORS['bull'], alpha=0.7, width=1, label='Bullish')
axes[1].bar(sample_s.index[bear_mask], sample_s.loc[bear_mask, 'NLP_Sentiment'],
            color=COLORS['bear'], alpha=0.7, width=1, label='Bearish')
axes[1].bar(sample_s.index[neut_mask], sample_s.loc[neut_mask, 'NLP_Sentiment'],
            color=COLORS['gray'], alpha=0.5, width=1, label='Neutral')
axes[1].set_title('Sentiment Classification Distribution')
axes[1].axhline(0, color=COLORS['white'], lw=0.7)
axes[1].legend(facecolor='#161b22', edgecolor='#30363d')
axes[1].grid(True, alpha=0.3)

# Sentiment vs Return scatter
axes[2].scatter(sample_s['NLP_Sentiment'], sample_s['Returns'],
                c=sample_s['NLP_Sentiment'], cmap='RdYlGn',
                alpha=0.4, s=8)
# Trend line
z = np.polyfit(sample_s['NLP_Sentiment'].dropna(),
               sample_s['Returns'].loc[sample_s['NLP_Sentiment'].dropna().index], 1)
p = np.poly1d(z)
x_line = np.linspace(-1, 1, 100)
axes[2].plot(x_line, p(x_line), color=COLORS['gold'], lw=2, label='Trend')
axes[2].axhline(0, color=COLORS['gray'], lw=0.7)
axes[2].axvline(0, color=COLORS['gray'], lw=0.7)
axes[2].set_title('Sentiment Score vs Daily Return (Correlation)')
axes[2].set_xlabel('NLP Sentiment Score')
axes[2].set_ylabel('Daily Return')
axes[2].legend(facecolor='#161b22', edgecolor='#30363d')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('08_nlp_sentiment.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('✅ NLP Sentiment Analysis complete')
```

```python
# ============================================================
# CELL 10: ENSEMBLE STRATEGY & TRADING SIGNALS
# ============================================================
print('🎯 Building Ensemble Trading Signal...')

# Align sentiment with test set
test_dates = df_feat.index[split_idx: split_idx + len(X_test)]
test_sentiment = df_feat.loc[test_dates, 'NLP_Sentiment'].values
# Normalize sentiment to 0-1 probability
sent_prob = (test_sentiment + 1) / 2

# Ensemble: RF (30%) + Boost (30%) + SVM (20%) + NLP Sentiment (20%)
ensemble_prob = (0.30 * rf_prob + 0.30 * boost_prob +
                 0.20 * svm_prob + 0.20 * sent_prob)

# Trading signals with confidence threshold
THRESHOLD_BUY  = 0.60
THRESHOLD_SELL = 0.40

signals = np.where(ensemble_prob > THRESHOLD_BUY, 1,
           np.where(ensemble_prob < THRESHOLD_SELL, -1, 0))

ensemble_pred = (ensemble_prob > 0.50).astype(int)
ens_acc = accuracy_score(y_test, ensemble_pred)
ens_auc = roc_auc_score(y_test, ensemble_prob)
print(f'   Ensemble Accuracy: {ens_acc:.4f} | AUC: {ens_auc:.4f}')
print(f'   BUY signals  : {(signals == 1).sum()}')
print(f'   SELL signals : {(signals == -1).sum()}')
print(f'   HOLD signals : {(signals == 0).sum()}')

# Backtesting
test_close = df_feat.loc[test_dates, 'Close'].values
daily_ret = np.diff(test_close) / test_close[:-1]
strategy_ret = signals[:-1] * daily_ret

cum_strategy = np.cumprod(1 + strategy_ret)
cum_buyhold = np.cumprod(1 + daily_ret)

# Sharpe Ratio
sharpe = (strategy_ret.mean() / (strategy_ret.std() + 1e-10)) * np.sqrt(252)
max_dd = np.min(cum_strategy / np.maximum.accumulate(cum_strategy) - 1)
total_return = cum_strategy[-1] - 1
bh_return = cum_buyhold[-1] - 1

print(f'\n📊 Backtest Results:')
print(f'   Strategy Return : {total_return:.2%}')
print(f'   Buy&Hold Return : {bh_return:.2%}')
print(f'   Sharpe Ratio    : {sharpe:.3f}')
print(f'   Max Drawdown    : {max_dd:.2%}')

# ── Ensemble Visualization ──
fig = plt.figure(figsize=(16, 14))
gs = gridspec.GridSpec(3, 2, figure=fig, hspace=0.4, wspace=0.3)
fig.suptitle('🎯 Ensemble Strategy & Backtesting Dashboard', fontsize=15, fontweight='bold', color=COLORS['white'])

# Cumulative Returns
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(cum_strategy, color=COLORS['bull'], lw=2.5, label=f'AI Strategy (+{total_return:.1%})')
ax1.plot(cum_buyhold, color=COLORS['blue'], lw=1.5, ls='--', label=f'Buy & Hold (+{bh_return:.1%})', alpha=0.8)
ax1.axhline(1, color=COLORS['gray'], lw=0.8, ls=':')
ax1.fill_between(range(len(cum_strategy)), cum_strategy, 1,
                  where=cum_strategy >= 1, alpha=0.2, color=COLORS['bull'])
ax1.fill_between(range(len(cum_strategy)), cum_strategy, 1,
                  where=cum_strategy < 1, alpha=0.2, color=COLORS['bear'])
ax1.set_title(f'Cumulative Returns: AI Strategy vs Buy & Hold | Sharpe: {sharpe:.2f} | Max DD: {max_dd:.1%}')
ax1.set_ylabel('Portfolio Value (×)')
ax1.legend(facecolor='#161b22', edgecolor='#30363d', fontsize=11)
ax1.grid(True, alpha=0.3)

# Signal distribution
ax2 = fig.add_subplot(gs[1, 0])
signal_counts = [np.sum(signals == 1), np.sum(signals == 0), np.sum(signals == -1)]
wedge_colors = [COLORS['bull'], COLORS['gray'], COLORS['bear']]
wedges, texts, autotexts = ax2.pie(signal_counts, labels=['BUY', 'HOLD', 'SELL'],
                                    colors=wedge_colors, autopct='%1.1f%%',
                                    startangle=90, pctdistance=0.8,
                                    wedgeprops={'linewidth': 2, 'edgecolor': '#0d1117'})
for t in autotexts:
    t.set_color(COLORS['white'])
    t.set_fontsize(11)
ax2.set_title('Trading Signal Distribution')

# Ensemble probability histogram
ax3 = fig.add_subplot(gs[1, 1])
ax3.hist(ensemble_prob[y_test == 1], bins=30, alpha=0.6, color=COLORS['bull'],
          label='Actual UP', density=True)
ax3.hist(ensemble_prob[y_test == 0], bins=30, alpha=0.6, color=COLORS['bear'],
          label='Actual DOWN', density=True)
ax3.axvline(THRESHOLD_BUY, color=COLORS['gold'], lw=1.5, ls='--', label=f'Buy Threshold {THRESHOLD_BUY}')
ax3.axvline(THRESHOLD_SELL, color=COLORS['pink'], lw=1.5, ls='--', label=f'Sell Threshold {THRESHOLD_SELL}')
ax3.set_title('Ensemble Probability Distribution')
ax3.set_xlabel('Probability Score')
ax3.legend(facecolor='#161b22', edgecolor='#30363d', fontsize=8)
ax3.grid(True, alpha=0.3)

# ROC all models + ensemble
ax4 = fig.add_subplot(gs[2, 0])
for name, prob, color in [('RF', rf_prob, COLORS['blue']),
                            (model_name[:3], boost_prob, COLORS['orange']),
                            ('SVM', svm_prob, COLORS['purple']),
                            ('Ensemble', ensemble_prob, COLORS['gold'])]:
    fpr, tpr, _ = roc_curve(y_test, prob)
    auc = roc_auc_score(y_test, prob)
    lw = 3 if name == 'Ensemble' else 1.5
    ax4.plot(fpr, tpr, color=color, lw=lw, label=f'{name} (AUC={auc:.3f})')
ax4.plot([0,1],[0,1], color=COLORS['gray'], lw=1, ls='--')
ax4.set_title('ROC: All Models + Ensemble')
ax4.set_xlabel('FPR')
ax4.set_ylabel('TPR')
ax4.legend(facecolor='#161b22', edgecolor='#30363d', fontsize=8)
ax4.grid(True, alpha=0.3)

# Summary scorecard
ax5 = fig.add_subplot(gs[2, 1])
ax5.axis('off')
scorecard = [
    ['Metric', 'Value', 'Rating'],
    ['Strategy Return', f'{total_return:.2%}', '✅' if total_return > bh_return else '❌'],
    ['Buy & Hold', f'{bh_return:.2%}', '📊'],
    ['Alpha Generated', f'{total_return-bh_return:.2%}', '⭐' if total_return > bh_return else '⬇️'],
    ['Sharpe Ratio', f'{sharpe:.3f}', '✅' if sharpe > 1 else '⚠️'],
    ['Max Drawdown', f'{max_dd:.2%}', '✅' if max_dd > -0.2 else '⚠️'],
    ['Ensemble AUC', f'{ens_auc:.4f}', '✅' if ens_auc > 0.55 else '⚠️'],
    ['RF Accuracy', f'{rf_acc:.2%}', '📈'],
    ['Total Signals', f'{len(signals):,}', '📊'],
]
row_colors = [['#21262d']*3] + [['#161b22', '#0d1117', '#161b22']]*(len(scorecard)-1)
table = ax5.table(cellText=scorecard[1:], colLabels=scorecard[0],
                   cellLoc='center', loc='center', bbox=[0, 0, 1, 1])
table.auto_set_font_size(False)
table.set_fontsize(10)
for (r, c), cell in table.get_celld().items():
    cell.set_facecolor('#21262d' if r == 0 else '#161b22')
    cell.set_text_props(color=COLORS['white'] if r > 0 else COLORS['gold'])
    cell.set_edgecolor('#30363d')
ax5.set_title('📊 Performance Scorecard', pad=12)

plt.savefig('09_ensemble_backtest.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('✅ Ensemble Strategy & Backtest complete')
```

```python
# ============================================================
# CELL 11: CROSS-VALIDATION & MODEL STABILITY
# ============================================================
print('🔬 Running Time-Series Cross-Validation...')

tscv = TimeSeriesSplit(n_splits=5)
cv_results = {}

models_cv = [
    ('Random Forest', rf_model),
    (model_name, boost_model),
    ('SVM', svm_model),
    ('Logistic Reg', lr_model)
]

for name, model in models_cv:
    scores = cross_val_score(model, X_train_sc, y_train, cv=tscv,
                              scoring='roc_auc', n_jobs=-1)
    cv_results[name] = scores
    print(f'   {name:20s}: {scores.mean():.4f} ± {scores.std():.4f}')

# ── CV Visualization ──
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('🔬 Time-Series Cross-Validation Stability', fontsize=14, fontweight='bold', color=COLORS['white'])

# CV scores per fold
model_colors_list = [COLORS['blue'], COLORS['orange'], COLORS['purple'], COLORS['teal']]
for i, (name, scores) in enumerate(cv_results.items()):
    axes[0].plot(range(1, 6), scores, marker='o', lw=2, ms=8,
                  color=model_colors_list[i], label=f'{name} (μ={scores.mean():.3f})')
axes[0].set_title('AUC-ROC per Cross-Validation Fold')
axes[0].set_xlabel('Fold')
axes[0].set_ylabel('AUC-ROC')
axes[0].set_xticks(range(1, 6))
axes[0].legend(facecolor='#161b22', edgecolor='#30363d', fontsize=9)
axes[0].grid(True, alpha=0.3)

# Box plot
bp_data = list(cv_results.values())
bp_labels = list(cv_results.keys())
bp = axes[1].boxplot(bp_data, labels=bp_labels, patch_artist=True,
                      medianprops={'color': COLORS['gold'], 'linewidth': 2},
                      whiskerprops={'color': COLORS['gray']},
                      capprops={'color': COLORS['gray']})
for patch, color in zip(bp['boxes'], model_colors_list):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)
axes[1].set_title('Model Stability Distribution')
axes[1].set_ylabel('AUC-ROC')
axes[1].tick_params(axis='x', rotation=10)
axes[1].grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('10_cross_validation.png', dpi=150, bbox_inches='tight', facecolor='#0d1117')
plt.show()
print('✅ Cross-Validation complete')
```
